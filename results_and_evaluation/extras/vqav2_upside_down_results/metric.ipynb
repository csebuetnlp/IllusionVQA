```python
import json
from matplotlib_venn import venn2
from matplotlib import pyplot as plt

question_ids_1 = set()
with open("dataset4_gemini_wrong.json", "r") as f:
    data = json.load(f)
    for i in range(len(data)):
        question_ids_1.add(data[i]["question_id"])

question_ids_2 = set()
with open("dataset4_gemini_inverted_wrong.json", "r") as f:
    data = json.load(f)
    for i in range(len(data)):
        question_ids_2.add(data[i]["question_id"])


venn2([question_ids_1, question_ids_2], set_labels = ('Gemini', 'Gemini Inverted'))
```
